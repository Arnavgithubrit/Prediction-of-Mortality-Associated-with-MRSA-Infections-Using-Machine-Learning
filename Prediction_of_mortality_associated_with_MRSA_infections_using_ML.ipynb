✅**Define Target**

```python
# Define target
target_col = 'MORTALITY YES=1,NO=0'

# Separate features and target
X = df.drop(columns=[target_col])
y = df[target_col]

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Perform train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 80% train, 20% test
    random_state=42,      # reproducibility
    stratify=y            # maintains class balance
)

# Check shapes
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

✅**5 Fold Cross Validation**

```python
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import make_scorer, precision_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier

# Fix column names
X_train.columns = X_train.columns.str.replace('[^A-Za-z0-9_]+', '_', regex=True)

# Safe precision scorer
precision_safe = make_scorer(precision_score, zero_division=0)

scoring = {
    'accuracy': 'accuracy',
    'precision': precision_safe,
    'recall': 'recall',
    'f1': 'f1',
    'roc_auc': 'roc_auc'
}

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, class_weight='balanced'),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced'),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "SVM": SVC(probability=True, class_weight='balanced'),
    "Decision Tree": DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    "XGBoost": XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42,
        n_estimators=100,
        max_depth=3
    )
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = {}

for name, model in models.items():
    cv_results = cross_validate(
        model,
        X_train,
        y_train,
        cv=skf,
        scoring=scoring,
        error_score='raise'
    )

    results[name] = {
        metric: cv_results['test_' + metric].mean()
        for metric in scoring
    }

# Print results
for model_name, metrics in results.items():
    print(f"\n🔹 {model_name}")
    for metric_name, value in metrics.items():
        print(f"{metric_name}: {value:.3f}")
```

✅**Bar graph for all models and all metrices across different models from the 5 fold cross validation**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Convert results to DataFrame
results_df = pd.DataFrame(results).T

models = results_df.index
metrics = results_df.columns

x = np.arange(len(models))
width = 0.12   # smaller width to avoid overlap

plt.figure(figsize=(12, 7))  # larger figure

# Plot grouped bars
for i, metric in enumerate(metrics):
    plt.bar(x + i*width, results_df[metric], width, label=metric)

# Improve spacing
plt.xlabel("Models", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.title("5-Fold Cross Validation Performance Comparison", fontsize=14)

# Better x-axis positioning
plt.xticks(x + width * (len(metrics)/2), models, rotation=30, ha='right')

# Move legend outside (VERY IMPORTANT)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Add grid for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Among all evaluated machine learning models, XGBoost demonstrated superior performance with the highest Accuracy(0.938), Precision(0.600) ROC-AUC (0.941), recall (0.50), and F1-score (0.533). Unlike other models such as Random Forest and KNN, which failed to identify mortality cases, XGBoost effectively captured clinically relevant outcomes, making it the most suitable model for mortality prediction

✅**ROC-AUC Curve for all the Validated Models**

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

roc_scores = {}

for name, model in models.items():

    # Generate cross-validated probability predictions
    y_prob = cross_val_predict(
        model,
        X_train,
        y_train,
        cv=skf,
        method="predict_proba"
    )[:, 1]

    # Compute ROC curve
    fpr, tpr, _ = roc_curve(y_train, y_prob)

    # Compute ROC-AUC
    auc_score = roc_auc_score(y_train, y_prob)

    roc_scores[name] = auc_score

    # Plot curve
    plt.plot(
        fpr,
        tpr,
        linewidth=2,
        label=f"{name} (AUC = {auc_score:.3f})"
    )


# Baseline reference line
plt.plot([0,1], [0,1], linestyle="--", linewidth=1)


# Labels
plt.xlabel("False Positive Rate", fontsize=14)
plt.ylabel("True Positive Rate", fontsize=14)
plt.title("5-Fold Cross-Validated ROC Curves for Mortality Prediction Models", fontsize=16)


# Legend
plt.legend(loc="lower right", fontsize=11)


# Grid
plt.grid(alpha=0.3)


# Save high-quality figure for publication
plt.savefig("CrossValidated_ROC_AUC_All_Models.png", dpi=600)


plt.show()


# Print ROC-AUC values again
print("\nCross-validated ROC-AUC scores:")
for model, score in roc_scores.items():
    print(f"{model}: {score:.3f}")
```

Cross-validated ROC curve analysis demonstrated that the XGBoost classifier achieved the highest discrimination ability (AUC = 0.961), outperforming all other evaluated machine learning models for mortality prediction
